```python
import networkx as nx
import matplotlib.pyplot as plt
import random 
from string import ascii_lowercase
```

```python
#city class struct
class Ville:
    def __init__(self,identifiant, x, y,name):
        self.identifiant = identifiant
        self.x = x
        self.y = y
        self.name = name
        
    def showDatas(self):
        print('Ville : id:',self.identifiant,'; X:',self.x,'; Y:',self.y)

#road class struct
class Route:
    def __init__(self,identifiant, id_ville_1, id_ville_2):
        self.identifiant = identifiant
        self.id_ville_1 = id_ville_1
        self.id_ville_2 = id_ville_2
        
    def showDatas(self):
        print('Route : id:',self.identifiant,'; Ville 1:',self.id_ville_1,'; Ville 2:',self.id_ville_2)
       
#defaults settings
nbr_villes = 10;
nbr_routes = 30;
maxX = 1000;
maxY = 1000;
minX = -1000;
minY = -1000;

#Lists
villes = []
routes = []
labels = {}
G = nx.Graph()


#Generate cities and associated nodes
for i in range(0,nbr_villes+1):
    x = random.randint(minX,maxX) #Need to add control to avoid generate same coords
    y = random.randint(minY,maxY)
    name = "test"
    ville = Ville(i,x,y,name)
    villes.append(ville)
    G.add_node(i,pos=(x,y))
    labels[i] = (str(i)+" "+ville.name)

#Generate roads and associated edges
for i in range(0,nbr_routes+1):
    id_ville_1 = random.randint(1,nbr_villes) #Need to add control to avoid generate same road 
    id_ville_2 = random.randint(1,nbr_villes)
    
    while id_ville_2 == id_ville_1:
        id_ville_2 = random.randint(1,nbr_villes)
    routes.append(Route(i,id_ville_1,id_ville_2))

#Draw Graph
pos = nx.get_node_attributes(G,'pos')
fig, ax = plt.subplots()        
nx.draw(G,pos,node_color='r', edge_color='b')
nx.draw_networkx_labels(G, pos, labels, font_size=16,font_weight='bold',ax=ax)
limits=plt.axis('on') # turns on axis
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)

#Output Cities datas values to debug
for ville in villes:
    ville.showDatas()
```
